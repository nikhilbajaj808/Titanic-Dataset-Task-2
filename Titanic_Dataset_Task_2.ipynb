```python
import pandas as pd

# ðŸ”„ Replace the path if your file is located somewhere else in Drive
file_path = '/content/drive/MyDrive/Titanic-Dataset.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
df.describe(include='all')  # includes both numeric and categorical
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for numeric features
df[['Age', 'Fare']].hist(bins=20, figsize=(12, 5), color='skyblue')
plt.suptitle("Histograms for Age and Fare")
plt.show()

# Plot boxplots
plt.figure(figsize=(12, 5))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplots for Age and Fare")
plt.show()
```

```python
# Use only numeric columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix
correlation = numeric_df.corr()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Numeric Features Only)")
plt.show()
```

```python
# Example: Group by survival
print(df.groupby('Survived')['Age'].mean())
print(df.groupby('Pclass')['Fare'].mean())

# Countplot for survival
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()

# Countplot for Pclass vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Pclass vs Survival")
plt.show()
```
